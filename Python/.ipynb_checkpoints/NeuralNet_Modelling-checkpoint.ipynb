```python

def CheckAccuracy(y_true, y_pred):
    stats = dict()
    stats['diff'] = y_true - y_pred
    stats['mse'] = np.mean(stats['diff']**2)
    print("Mean Squared Error:   " , stats['mse'])
    stats['mape'] = np.mean(np.abs(stats['diff'] / y_true)) 
    print("Mean Absolute Percentage Error:   " , stats['mape'])
    return stats

# import data
df = pd.read_csv('C:/Users/User/Desktop/Data speciale/NeuralNet_data.csv', parse_dates= True, index_col=0)
```

```python



df.head()
```

```python

df = df.sort_index()


unique_dates = df.index.unique()


split_index = int(len(unique_dates) * 0.95)
train_dates = unique_dates[:split_index] 
test_dates = unique_dates[split_index:]   

# Split the dataset by the determined dates
X_train = df.loc[train_dates, ['Stock (S)', 'Strike (K)', 'Moneyness (M)', 'Time to maturity (T)', 
                               'Risk free rate (r)', 'BS (vol10)', 'BS (vol30)', 'BS (vol60)', 
                               'vol10', 'vol30', 'vol60', 'Year']].values
X_test = df.loc[test_dates, ['Stock (S)', 'Strike (K)', 'Moneyness (M)', 'Time to maturity (T)', 
                             'Risk free rate (r)', 'BS (vol10)', 'BS (vol30)', 'BS (vol60)', 
                             'vol10', 'vol30', 'vol60', 'Year']].values

y_train = df.loc[train_dates, ['Option Price (C)']].values
y_test = df.loc[test_dates, ['Option Price (C)']].values
strike_train = X_train[:, 1]
year_train = X_train[:, -1]
BSM_train = X_train[:, 5:8]
vol_train = X_train[:, 9]
strike_test = X_test[:, 1]
year_test = X_test[:, -1]
BSM_test = X_test[:, 5:8]
vol_test = X_test[:, 9]
X_train = X_train[:, 0:5]
X_test = X_test[:, 0:5]


```

```python

# Scale the data
scaler = RobustScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
vol_train = scaler.fit_transform(vol_train.reshape(-1, 1)).flatten()
vol_test = scaler.transform(vol_test.reshape(-1, 1)).flatten()
X_train_scaled = np.column_stack((X_train, vol_train))
X_test_scaled = np.column_stack((X_test, vol_test))
y_train_scaled = scaler.fit_transform(y_train)
y_test_scaled = scaler.transform(y_test)

```

```python
# Create the model
model = Sequential()
model.add(Dense(32, input_dim=6))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train_scaled, y_train_scaled, epochs=50, batch_size=4096, verbose=True)




```

```python
y_train_hat = y_train_scaled.invese_transform(model.predict(X_train_scaled))

y_test_hat = y_test_scaled.invese_transform(model.predict(X_test_scaled))
```
