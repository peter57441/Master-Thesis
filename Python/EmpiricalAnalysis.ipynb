```python

def CheckAccuracy(y_true, y_pred):
    stats = dict()
    stats['diff'] = y_true - y_pred
    stats['rmse'] = np.sqrt(np.mean(stats['diff']**2))
    print("Root Mean Squared Error:   " , stats['rmse'])
    stats['mape'] = np.mean(np.abs(stats['diff'] / y_true)) 
    print("Mean Absolute Percentage Error:   " , stats['mape'])
    stats['mse'] = np.mean(stats['diff']**2)
    print("Mean Squared Error:   " , stats['mse'])
    stats['mae'] = np.mean(np.abs(stats['diff']))
    print("Mean Absolute Error:   " , stats['mae'])
    return stats


# Check accuracy of the model

print("Accuracy of the ANN model on the test set")
print(" ")
print("VIX")
stats = CheckAccuracy(df_test['Option Price (C)'], df_test['ANN VIX'])
print(" ")
print("GARCH")
stats = CheckAccuracy(df_test['Option Price (C)'], df_test['ANN GARCH'])
print(" ")
print("vol60")
stats = CheckAccuracy(df_test['Option Price (C)'], df_test['ANN vol60'])
print(" ")
print("vol30")
stats = CheckAccuracy(df_test['Option Price (C)'], df_test['ANN vol30'])
print(" ")
print("vol10")
stats = CheckAccuracy(df_test['Option Price (C)'], df_test['ANN vol10'])

print(" ")
print("Accuary of the BSM model on the test set")
print(" ")
print("VIX")
stats = CheckAccuracy(df_test['Option Price (C)'], df_test['BS VIX'])
print(" ")
print("GARCH")
stats = CheckAccuracy(df_test['Option Price (C)'], df_test['BS GARCH'])
print(" ")
print("vol60")
stats = CheckAccuracy(df_test['Option Price (C)'], df_test['BS (vol60)'])
print(" ")
print("vol30")
stats = CheckAccuracy(df_test['Option Price (C)'], df_test['BS (vol30)'])
print(" ")
print("vol10")
stats = CheckAccuracy(df_test['Option Price (C)'], df_test['BS (vol10)'])
```

# Time-to-Maturity sensitivity analysis

```python

# partition the test_set into 5 bins based on the time to maturity of the options and calculate the RMSE, MAE and MAPE for each bin for both the ANN and BSM models
# the first bin should be 0-14 days, the second 15-28 days, the third 29-90 days, the fourth 91-180 days and the fifth 181 days and above
# the time to maturity column is the difference between the expiration date and the current date / 365

# 0-14 days
bin1 = df_test[df_test['Time to maturity (T)'] <= 14/365]
# 15-28 days
bin2 = df_test[(df_test['Time to maturity (T)'] > 14/365) & (df_test['Time to maturity (T)'] <= 28/365)]
# 29-90 days
bin3 = df_test[(df_test['Time to maturity (T)'] > 28/365) & (df_test['Time to maturity (T)'] <= 90/365)]
# 91-180 days
bin4 = df_test[(df_test['Time to maturity (T)'] > 90/365) & (df_test['Time to maturity (T)'] <= 180/365)]
# 181 days and above
bin5 = df_test[df_test['Time to maturity (T)'] > 180/365]

# calculate the RMSE, MAE and MAPE for each bin for both the ANN and BSM models and save the results in a latex table

# ANN VIX
RMSE_bin1 = np.sqrt(np.mean((bin1['Option Price (C)'] - bin1['ANN VIX'])**2))
RMSE_bin2 = np.sqrt(np.mean((bin2['Option Price (C)'] - bin2['ANN VIX'])**2))
RMSE_bin3 = np.sqrt(np.mean((bin3['Option Price (C)'] - bin3['ANN VIX'])**2))
RMSE_bin4 = np.sqrt(np.mean((bin4['Option Price (C)'] - bin4['ANN VIX'])**2))
RMSE_bin5 = np.sqrt(np.mean((bin5['Option Price (C)'] - bin5['ANN VIX'])**2))

MAE_bin1 = np.mean(np.abs(bin1['Option Price (C)'] - bin1['ANN VIX']))
MAE_bin2 = np.mean(np.abs(bin2['Option Price (C)'] - bin2['ANN VIX']))
MAE_bin3 = np.mean(np.abs(bin3['Option Price (C)'] - bin3['ANN VIX']))
MAE_bin4 = np.mean(np.abs(bin4['Option Price (C)'] - bin4['ANN VIX']))
MAE_bin5 = np.mean(np.abs(bin5['Option Price (C)'] - bin5['ANN VIX']))

MAPE_bin1 = np.mean(np.abs((bin1['Option Price (C)'] - bin1['ANN VIX']) / bin1['Option Price (C)']))
MAPE_bin2 = np.mean(np.abs((bin2['Option Price (C)'] - bin2['ANN VIX']) / bin2['Option Price (C)']))
MAPE_bin3 = np.mean(np.abs((bin3['Option Price (C)'] - bin3['ANN VIX']) / bin3['Option Price (C)']))
MAPE_bin4 = np.mean(np.abs((bin4['Option Price (C)'] - bin4['ANN VIX']) / bin4['Option Price (C)']))
MAPE_bin5 = np.mean(np.abs((bin5['Option Price (C)'] - bin5['ANN VIX']) / bin5['Option Price (C)']))

# BSM vol60

RMSE_bin1_BSM = np.sqrt(np.mean((bin1['Option Price (C)'] - bin1['BS (vol60)'])**2))
RMSE_bin2_BSM = np.sqrt(np.mean((bin2['Option Price (C)'] - bin2['BS (vol60)'])**2))
RMSE_bin3_BSM = np.sqrt(np.mean((bin3['Option Price (C)'] - bin3['BS (vol60)'])**2))
RMSE_bin4_BSM = np.sqrt(np.mean((bin4['Option Price (C)'] - bin4['BS (vol60)'])**2))
RMSE_bin5_BSM = np.sqrt(np.mean((bin5['Option Price (C)'] - bin5['BS (vol60)'])**2))

MAE_bin1_BSM = np.mean(np.abs(bin1['Option Price (C)'] - bin1['BS (vol60)']))
MAE_bin2_BSM = np.mean(np.abs(bin2['Option Price (C)'] - bin2['BS (vol60)']))
MAE_bin3_BSM = np.mean(np.abs(bin3['Option Price (C)'] - bin3['BS (vol60)']))
MAE_bin4_BSM = np.mean(np.abs(bin4['Option Price (C)'] - bin4['BS (vol60)']))
MAE_bin5_BSM = np.mean(np.abs(bin5['Option Price (C)'] - bin5['BS (vol60)']))

MAPE_bin1_BSM = np.mean(np.abs((bin1['Option Price (C)'] - bin1['BS (vol60)']) / bin1['Option Price (C)']))
MAPE_bin2_BSM = np.mean(np.abs((bin2['Option Price (C)'] - bin2['BS (vol60)']) / bin2['Option Price (C)']))
MAPE_bin3_BSM = np.mean(np.abs((bin3['Option Price (C)'] - bin3['BS (vol60)']) / bin3['Option Price (C)']))
MAPE_bin4_BSM = np.mean(np.abs((bin4['Option Price (C)'] - bin4['BS (vol60)']) / bin4['Option Price (C)']))
MAPE_bin5_BSM = np.mean(np.abs((bin5['Option Price (C)'] - bin5['BS (vol60)']) / bin5['Option Price (C)']))

rows = [

    ['Time to Maturity', 'Model', 'RMSE', 'MAE', 'MAPE'],
    ['0-14 days', 'ANN', round(RMSE_bin1, 5), round(MAE_bin1, 5), round(MAPE_bin1, 5)],
    ['', 'BSM', round(RMSE_bin1_BSM, 5), round(MAE_bin1_BSM, 5), round(MAPE_bin1_BSM, 5)],
    ['', '', '', '', ''],
    ['15-28 days', 'ANN', round(RMSE_bin2, 5), round(MAE_bin2, 5), round(MAPE_bin2, 5)],
    ['', 'BSM', round(RMSE_bin2_BSM, 5), round(MAE_bin2_BSM, 5), round(MAPE_bin2_BSM, 5)],
    ['', '', '', '', ''],
    ['29-90 days', 'ANN', round(RMSE_bin3, 5), round(MAE_bin3, 5), round(MAPE_bin3, 5)],
    ['', 'BSM', round(RMSE_bin3_BSM, 5), round(MAE_bin3_BSM, 5), round(MAPE_bin3_BSM, 5)],
    ['', '', '', '', ''],
    ['91-180 days', 'ANN', round(RMSE_bin4, 5), round(MAE_bin4, 5), round(MAPE_bin4, 5)],
    ['', 'BSM', round(RMSE_bin4_BSM, 5), round(MAE_bin4_BSM, 5), round(MAPE_bin4_BSM, 5)],
    ['', '', '', '', ''],
    ['181 days and above', 'ANN', round(RMSE_bin5, 5), round(MAE_bin5, 5), round(MAPE_bin5, 5)],
    ['', 'BSM', round(RMSE_bin5_BSM, 5), round(MAE_bin5_BSM, 5), round(MAPE_bin5_BSM, 5)],
    ['', '', '', '', ''],
    ['All', 'ANN',
     round(np.sqrt(np.mean((df_test['Option Price (C)'] - df_test['ANN VIX']) ** 2)), 5),
     round(np.mean(np.abs(df_test['Option Price (C)'] - df_test['ANN VIX'])), 5),
     round(np.mean(np.abs((df_test['Option Price (C)'] - df_test['ANN VIX']) / df_test['Option Price (C)'])), 5)
    ],
    ['', 'BSM',
     round(np.sqrt(np.mean((df_test['Option Price (C)'] - df_test['BS (vol60)']) ** 2)), 5),
     round(np.mean(np.abs(df_test['Option Price (C)'] - df_test['BS (vol60)'])), 5),
     round(np.mean(np.abs((df_test['Option Price (C)'] - df_test['BS (vol60)']) / df_test['Option Price (C)'])), 5)
    ]
]

print('---------------------- All samples ---------------------------')
print('Tabulate Table:')
print(tabulate(rows, headers='firstrow', tablefmt='latex'))


```

# Volatility Surface sensitivity analysis

```python
# Define bins for time to maturity and moneyness
time_bins = [
    (0, 14 / 365),
    (14 / 365, 28 / 365),
    (28 / 365, 90 / 365),
    (90 / 365, 180 / 365),
    (180 / 365, 730 / 365)
]
moneyness_bins = [
    (0.8, 0.9),
    (0.9, 0.98),
    (0.98, 1.02),
    (1.02, 1.1),
    (1.1, 1.5)
]

# Create a 5x5 grid to hold RMSE values
rmse_grid = np.zeros((len(time_bins), len(moneyness_bins)))

# Compute RMSE for each bin
for i, (t_min, t_max) in enumerate(time_bins):
    for j, (m_min, m_max) in enumerate(moneyness_bins):
        bin_data = df_test[
            (df_test['Time to maturity (T)'] > t_min) & (df_test['Time to maturity (T)'] <= t_max) &
            (df_test['Moneyness (M)'] >= m_min) & (df_test['Moneyness (M)'] < m_max)
        ]
        if not bin_data.empty:  # Avoid computing RMSE for empty bins
            rmse = np.sqrt(np.mean((bin_data['Option Price (C)'] - bin_data['ANN VIX']) ** 2))
            rmse_grid[i, j] = rmse
        else:
            rmse_grid[i, j] = np.nan  # Mark empty bins as NaN for better visualization

# Set up the heatmap
plt.figure(figsize=(10, 6))
ax = sns.heatmap(
    rmse_grid,
    annot=True,
    fmt=".5f",
    cmap="RdYlGn_r",
    linewidths=0.5,
    xticklabels=[['DOTM', 'OTM', 'ATM', 'ITM', 'DITM'][m] for m in range(len(moneyness_bins))],
    yticklabels=[['Very Short', 'Short', 'Medium', 'Long', 'Very Long'][i] for i in range(len(time_bins))
    ],
    cbar_kws={"label": "RMSE"}
)

# Customize labels and title
ax.set_title("Artificial Neural Network", fontsize=14)
ax.set_xlabel("Moneyness", fontsize=12)
ax.set_ylabel("Time-To-Maturity", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

plt.show()
```
